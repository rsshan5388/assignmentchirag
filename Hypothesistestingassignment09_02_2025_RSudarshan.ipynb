7.**[Question from previous assignment]**  A researcher wants to estimate the average height of students in a university. She samples
50 students and finds the mean height to be 65 inches with a standard deviation of 3 inches.
Construct a 95% confidence interval for the population mean height


**Answer and Explanation:**

The formula for confidence interval is $CI = \bar{x} \pm t \times \frac{s}{\sqrt{n}}$

where:
$\bar{x}$ = 65 inch(sample mean)

s= 3 inch (sample standard deviation)

n=50 (sample size)

t is the critical value from the t-distribution with n-1=49 degrees of freedom at a 95% confidence level.

**Step 1: Find the critical t-value**

For a 95% confidence level and 49 degrees of freedom, the critical t-value from a t-table or statistical calculator is approximately 2.009.

**Step 2: Compute the margin of error (ME)**

$ME =  t \times \frac{s}{\sqrt{n}}=2.009 \times \frac{3}{\sqrt{50}}$

ME=0.8528

**Step 3: Construct the Confidence Interval**

**CI=65¬±0.853**

CI=(64.15,65.85)

**Conclusion: **

With 95% confidence, the true average height of all students at the university is estimated to be between 64.15 inches and 65.85 inches.









```python
#Calculation using python code
sample_mean = 65  # Sample mean height
sample_std = 3     # Sample standard deviation
sample_size = 50   # Sample size
confidence_level = 0.95

# Degrees of freedom
df = sample_size - 1

# Critical t-value for 95% confidence level
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

# Margin of error (ME)
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Confidence Interval (CI)
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print results
print("\033[1mExpected Output\033[0m")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

8.**[Question from previous assignment]** A manufacturer claims that the average lifespan of its light bulbs is 1000 hours. A random
sample of 50 light bulbs has a mean lifespan of 980 hours with a standard deviation of 50 hours.
Test the manufacturer's claim at a significance level of 0.05 using a right-tailed hypothesis test

Ans:

**Step 1: Define Hypotheses **

Null Hypothesis ${H_0}$= The average lifespan of the light bulbs is 1000 hours.

${H_0}$:ùúá=1000

Alernate Hypothesis ${H_A}$ > 1000

**Step 2: Given Data**

Sample size: n = 50

Sample mean: $\bar x$=980

Population mean (claimed): ùúá=1000

Sample standard deviation: s=50

Significance level: ùõº=0.05

**Step 3: Compute Test Statistic**

We use the t-test because the population standard deviation is unknown. The formula for the t-score is:

$$ t = (\bar{x} - \mu) \div (s / \sqrt{n}) $$

$$ t =(980 -1000) \div (50/\sqrt{50}) $$
$$t = (-20) \div (50/7.071) $$
$$ t=  -2.828 $$

**Step 4: Find the Critical Value**
Since this is a right-tailed test at a 0.05 significance level, we find the critical t-value for df = n-1=49

The critical t-value is found using a t-table or using Python.

Degrees of Freedom (df): 49

Significance Level (ùõº):0.05

Right-Tailed Test means we look for ùë°ùõº  where (ùëá>ùë°ùõº)=0.05

From a t-table, the critical value for df = 49 at ùõº =0.05 (right-tailed test) is 1.677

**Step 5: Compare t-Statistic and Critical Value**

Our computed t-statistic: -2.828

Critical t-value:1.677

1.677 is greater than -2.828  we fail to reject the null hypothesis.

**Step 5: Compute the p-Value**

The p-value for a right-tailed test is: P=1‚àíùëÉ(ùëá‚â§ùë°satistic)
The p-value (0.997) is much greater than  ùõº =0.05
 meaning there is no statistical evidence to support the claim that the lifespan is significantly greater than 1000 hours.









```python
sample_mean = 980  # Sample mean lifespan
sample_std = 50     # Sample standard deviation
sample_size = 50    # Sample size
claimed_mean = 1000 # Claimed population mean
alpha = 0.05        # Significance level

# Degrees of freedom
df = sample_size - 1

# Compute t-statistic
t_statistic = (sample_mean - claimed_mean) / (sample_std / np.sqrt(sample_size))

# Find critical t-value for right-tailed test (alpha = 0.05)
t_critical = stats.t.ppf(1 - alpha, df)

# Compute p-value for right-tailed test
p_value = 1 - stats.t.cdf(t_statistic, df)

# Print results
print(f"T-Statistic: {t_statistic:.3f}")
print(f"Critical T-Value: {t_critical:.3f}")
print(f"P-Value: {p_value:.3f}")

# Decision Rule
if t_statistic > t_critical:
    print("Reject the null hypothesis: The average lifespan is significantly greater than 1000 hours.")
else:
    print("Fail to reject the null hypothesis: No evidence to suggest the average lifespan is greater than 1000 hours.")
```

# 4. Type I and Type II Errors:
- Describe a scenario in which a Type I error could occur in hypothesis testing. How
does it differ from a Type II error?

Ans:
In hypothesis testing, we make decisions based on sample data, and sometimes, we make errors. The two types of errors are:

**Type I Error (False Positive) **‚Üí Rejecting a true null hypothesis ($ùêª_0$)
Type II Error (False Negative) ‚Üí Failing to reject a false null hypothesis ($ùêª_0$)

**Scenario for a Type I Error (False Positive)**

Example: COVID-19 Testing
Null Hypothesis ($ùêª_0$): The person does not have COVID-19.

Alternative Hypothesis ($ùêª_ùê¥$): The person has COVID-19.
 Type I Error occurs if the test incorrectly detects COVID-19 in a healthy person (false positive).

Reality: The person is healthy (no COVID-19).
Test Result: The test says they have COVID-19 (false positive).

Consequence: The person may be unnecessarily quarantined, given unneeded treatment, and experience stress.

**Scenario for a Type II Error (False Negative)**

Using the same COVID-19 testing example:

Type II Error occurs if the test fails to detect COVID-19 in an infected person (false negative).

Reality: The person actually has COVID-19.

Test Result: The test says they do not have COVID-19 (false negative).

 Consequence: The infected person may unknowingly spread the virus, and not receive needed treatment, leading to health risks.

**How to Reduce These Errors?**
Lowering Œ± (significance level) reduces Type I error but increases Type II error.

Increasing sample size helps balance both errors.

Using more accurate tests (better measurement tools) reduces both errors.

**Conclusion:**Type I error is like a false alarm (detecting something that isn‚Äôt there). Type II error is like missing a real problem (failing to detect something that exists)









# 5. Right-tailed Hypothesis Testing:
- A manufacturer claims that their new light bulb lasts, on average, more than 1000
hours. Conduct a right-tailed hypothesis test with a significance level of 0.05, given a
sample mean of 1050 hours and a sample standard deviation of 50 hours.

Ans:
We are given:

Claimed Mean Lifespan: $ùúá_0$ =1000 hours

Sample Mean: $\bar x$=1050 hours

Sample Standard Deviation:s=50 hours

Sample Size:n=30

(assuming) Significance Level: Œ±=0.05

Test Type: Right-tailed t-test (since we are testing if the mean is greater than 1000 hours)

**Step 1: Define Hypotheses**
**Null Hypothesis** ($ùêª_0$): The average lifespan of the light bulbs is 1000 hours or less.

$H_0$:Œº=1000

**Alternative Hypothesis** ($H_A$): The average lifespan is greater than 1000 hours.

$ùêª_A$:ùúá>1000

Since we are testing if the lifespan increases, this is a right-tailed test.

**Step 2: Compute the Test Statistic**

The t-score formula: $t=(\bar x -\mu_0) \div (s /\sqrt(n))$

Substituting values:

$t=(1050-1000) \div (50/\sqrt(30))$

$t=(50) \div (50/5.477)$

$t=5.477$

**Step 3: Find the Critical t-Value **

Since this is a right-tailed test at $\alpha$ = 0.05 , we find the critical t-value for df=30 -1=29

Using a t-table or scipy.stats.t.ppf(), we find:$ {ùë°}_{critical}$=1.699

**Step 4: Compare t-Statistic and Critical Value**
Our computed t-statistic: 5.477

Critical t-value: 1.699
Since t=5.477 > ${t}_{critical}$=1.699 so  we reject the null hypothesis.

**Step 5: Compute the p-Value**
The p-value for a right-tailed test is:

p=1 - P(T<tstatistic )

p-value from table ~ 0.000

p-value < 0.05 so its confirmatory of rejecting null hypothesis

**Conclusion: A**t a 0.05 significance level, we have strong statistical evidence to reject the null hypothesis. This means the new light bulb has a significantly greater lifespan than 1000 hours



















```python
sample_mean = 1050  # Sample mean lifespan
sample_std = 50      # Sample standard deviation
sample_size = 30     # Sample size
claimed_mean = 1000  # Claimed population mean
alpha = 0.05         # Significance level

# Degrees of freedom
df = sample_size - 1

# Compute t-statistic
t_statistic = (sample_mean - claimed_mean) / (sample_std / np.sqrt(sample_size))

# Find critical t-value for right-tailed test (alpha = 0.05)
t_critical = stats.t.ppf(1 - alpha, df)

# Compute p-value for right-tailed test
p_value = 1 - stats.t.cdf(t_statistic, df)

# Print results
print(f"T-Statistic: {t_statistic:.3f}")
print(f"Critical T-Value: {t_critical:.3f}")
print(f"P-Value: {p_value:.6f}")

# Decision Rule

if t_statistic > t_critical:
    print("Reject the null hypothesis: The average lifespan is significantly greater than 1000 hours.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the average lifespan is greater than 1000 hours.")
```

# 7. One-sample t-test:
A manufacturer claims that the mean weight of their cereal boxes is 500 grams. A
sample of 30 cereal boxes has a mean weight of 490 grams and a standard deviation of
20 grams. Conduct a one-sample t-test to determine if there is evidence to support the
manufacturer's claim at a significance level of 0.05.

Ans:
Given

Sample Mean $\bar x$ = 490 grams

Sample Standard Deviation s = 20 grams

Sample Size n = 30

Significance Level Œ± = 0.05

We will conduct a one-sample t-test to determine whether there is sufficient evidence to reject the manufacturer's claim.

**Step 1: Define the Hypotheses**

Null Hypothesis $H_0$: The mean weight of the cereal boxes is 500 grams (as claimed by the manufacturer. ùúá=500

Alternative Hypothesis $H_A$: The mean weight is different from 500 grams.

Since we are testing for any difference (either higher or lower), this is a two-tailed test.

Step 2: Determine the Test Statistic

We use the t-test formula for a one-sample test:

$t= (\bar x - \mu_0) \div (s/\sqrt(n))$

Substituting the given values:
$t=(490 - 500) \div (20/\sqrt(30))$

$t=(-10) \div (20/5.477)$

$t=-2.739$

Thus, our computed t-statistic is -2.739.

**Step 3: Find the Critical t-Value**

Since this is a two-tailed test, we need the critical t-value for

Significance level (Œ± = 0.05)
Degrees of freedom (df) = n-1 =30 -1=29

Using a t-table or Python, the critical t-value for a two-tailed test at
Œ±=0.05 and df = 29 is:

${t}_{critical}$=¬±2.045

**Step 4: Compare t-Statistic with Critical Value**

Computed t-statistic: -2.739

Critical t-values: ¬±2.045

Since:
${t}_{statistic}$ < ${t}_{critical}$

(i.e.,-2.739<-2.045)

We reject the null hypothesis.

**Step 5: Compute the p-Value**

The p-value represents the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true.

For a two-tailed test, the p-value is calculated as:

$p=2 \times P(T<‚àí2.739) $

Using Python, we found: $ùëù = 0.0104 $
Since $p=0.0104<Œ±=0.05$, we have sufficient evidence to reject $ùêª_0$

**Step 6: Interpretation & Conclusion**
Since our computed t-statistic (-2.739) is beyond the critical t-value (-2.045) and our p-value (0.0104) is less than 0.05, we reject the null hypothesis.

Conclusion:
At a 0.05 significance level, we have strong statistical evidence that the mean weight of the cereal boxes is significantly different from 500 grams.

This means that the manufacturer's claim is likely incorrect, and further investigation may be needed to check underfilling or overfilling issues in production.













```python
import scipy.stats as stats
import numpy as np

# Given data
sample_mean = 490  # Sample mean weight
sample_std = 20     # Sample standard deviation
sample_size = 30    # Sample size
claimed_mean = 500  # Claimed population mean
alpha = 0.05        # Significance level

# Degrees of freedom
df = sample_size - 1

# Compute t-statistic
t_statistic = (sample_mean - claimed_mean) / (sample_std / np.sqrt(sample_size))

# Find critical t-value for two-tailed test (alpha = 0.05)
t_critical = stats.t.ppf(1 - alpha/2, df)  # Two-tailed test

# Compute p-value for two-tailed test
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

# Print results
print(f"T-Statistic: {t_statistic:.3f}")
print(f"Critical T-Value: ¬±{t_critical:.3f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule
if abs(t_statistic) > t_critical:
    print("Reject the null hypothesis: The mean weight is significantly different from 500 grams.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the mean weight is different from 500 grams.")
```

Standard deviation $s_A$ = 0.1 seconds

Sample size $ùëõ_A$ = 30

Group B:

Mean reaction time $\bar x_B$= 0.55 seconds

Standard deviation $s_B$ = 0.08 seconds

Sample size $n_B$=30

Significance Level: Œ±=0.01


Test Type: Two-sample t-test (Welch‚Äôs t-test, assuming unequal variances).

**Step 1: Define the Hypotheses**

Null Hypothesis ( $H_0$ ): There is no significant difference in the mean reaction times between Group A and Group B.

$H_0$ : $Œº_A$ =$Œº_B$

or equivalently,  $H_0$ : $Œº_A$ - $Œº_B$=0

where: Œº1 = Mean exam score of Group 1.

Œº2 = Mean exam score of Group 2.

Alternative Hypothesis ( $H_A$ ): There is a significant difference in the mean exam scores between the two groups (either higher or lower).  $H_A$ : $Œº_A$ ‚â† $Œº_B$

or equivalently,  $H_A$ : $Œº_A$  -  $Œº_B$ ‚â†0

**Step 2: Compute the Test Statistic**

$t= \frac{\bar x_A - \bar x_B} {\sqrt{(s_A)^2/n_A + (s_B)^2/n_B}}$


Substituting the given values:

$t=\frac{0.6 - 0.55} {\sqrt{(0.1)^2/30 + (0.08)^2/30}}$

$t= \frac{0.05} {\sqrt{(0.01/30 + (0.0064)/30}}$

$t= \frac{0.05} {\sqrt{0.0003333 + 0.0002133}}$

$t= \frac{0.05} {\sqrt{0.0005467}}$

$t= \frac{0.05} {0.02338}$

$t=2.138$

Thus, our computed $t_{statistic}$ is 2.138.

**Step 3: Find the Critical t-Value**

Since this is a two-tailed test at Œ±=0.01, we need to find the critical t-value for our degrees of freedom.

Degrees of Freedom (Welch‚Äôs Formula)
Since the sample variances are not assumed to be equal, we use Welch‚Äôs approximation for degrees of freedom

$df = \frac{(s_A ^2/n_A \dotplus  s_B^2/n_B)^2}{(s_A^2/n_A)^2/n_A -1)+(s_B^2/n_B)^2/n_B -1)}$

Substituting values:

$df = \frac{(0.0005467)^2}{(0.0003333)^2/29)+(0.0002133)^2/29)}$

$df \approx 55.33$

Using a t-table or Python, the critical  t-value for $df \approx 55.33 $

at Œ±=0.01 (two-tailed) is:$ \pm 2.668 $

**Step 4: Compare t-Statistic with Critical Value**

Computed t-statistic: 2.138

Critical t-values: ¬±2.668

Since:

t =2.138 < $t_{critical}$=2.668
we fail to reject the null hypothesis.

**Step 5: Compute the p-Value**
The p-value represents the probability of obtaining a test statistic as extreme as the observed one, assuming $H_0$ is true.

For a two-tailed test, the p-value is calculated as:

$P= 2 \times (P > \big|{t}\big|)$

$P=0.0369$

Since p=0.0369>Œ±=0.01, we do not have enough evidence to reject $H_0$

Since $\big|ùë°\big|$ =2.138  is less than  2.668 and p-value (0.0369) is greater than 0.01, we fail to reject the null hypothesis.

**Final Conclusion:**
At a 0.01 significance level, we do not have sufficient evidence to conclude that the mean reaction times of Group A and Group B are significantly different.

Although there is some difference in the means (0.6 vs. 0.55 seconds), this difference is not statistically significant at the 1% level.








```python
mean_A = 0.6   # Mean reaction time (seconds)
std_A = 0.1    # Standard deviation (seconds)
n_A = 30       # Sample size

# Given data for Group B
mean_B = 0.55  # Mean reaction time (seconds)
std_B = 0.08   # Standard deviation (seconds)
n_B = 30       # Sample size

# Significance level
alpha = 0.01

# Compute t-statistic for two-sample independent t-test (Welch's t-test for unequal variances)
t_statistic, p_value = stats.ttest_ind_from_stats(mean1=mean_A, std1=std_A, nobs1=n_A,
                                                  mean2=mean_B, std2=std_B, nobs2=n_B,
                                                  equal_var=False)  # Welch's t-test

# Compute degrees of freedom using Welch‚Äôs formula
df = ( (std_A**2/n_A + std_B**2/n_B)**2 ) / \
     ( (std_A**2/n_A)**2 / (n_A - 1) + (std_B**2/n_B)**2 / (n_B - 1) )

# Find critical t-value for two-tailed test
t_critical = stats.t.ppf(1 - alpha/2, df)

# Print results
print(f"T-Statistic: {t_statistic:.3f}")
print(f"Degrees of Freedom: {df:.2f}")
print(f"Critical T-Value (Two-Tailed): ¬±{t_critical:.3f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule
if abs(t_statistic) > t_critical:
    print("Reject the null hypothesis: There is a significant difference in reaction times between Group A and Group B.")
else:
    print("Fail to reject the null hypothesis: No significant difference in reaction times between Group A and Group B.")
```

# 9. Process Control Example:
- A call center manager implements a new training program aimed at reducing call
waiting times. The average waiting time before the training program was 4.5 minutes, and
after the program, it is measured to be 4.0 minutes with a standard deviation of 0.8
minutes. Conduct a hypothesis test to determine if there is evidence that the training
program has reduced waiting times, using a significance level of 0.05.

Ans:

Here NULL Hypothesis $H_0$ is training program did not reduce waiting time below 4.5 minutes

i.e $H_0 : \mu =4.5 minutes$

Alernate Hypothesis $H_A$ is the traininh program reduced the waiting time belwow 4.5 minutes

i.e $H_A : \mu < 4.5 minutes$

so we carry out a left-tailed test because we are specifically looking for a decrease in waiting times.

formula

$t=\frac{\bar x - \mu_o }{s/\sqrt(n)}$

Substituting values:
$t=\frac{4 -4.5}{0.8/\sqrt(30)}$

$t=\frac{-0.5}{0.8/5.477}$

$t=-3.423$

This is our $t_{statisitics}$=-3.423
The $t_{critical}$ for $\alpha$=0.05 with df =29 is -1.699

P-value here for p =P(T <  $t_{statisitics}$) is 0.0009 where is very much less than $\alpha$=0.05 , so we reject NULL Hypothesis $H_0$

**Conclusion:**

Since the t-statistic is more extreme than the critical value and the p-value is very small, we have strong statistical evidence to reject the null hypothesis.

At a 0.05 significance level, we conclude that the training program significantly reduced call waiting times.













```python
# Given data
before_mean = 4.5  # Mean waiting time before training
after_mean = 4.0   # Mean waiting time after training
after_std = 0.8    # Standard deviation after training
n = 30             # Sample size
alpha = 0.05       # Significance level

# Compute t-statistic
t_statistic = (after_mean - before_mean) / (after_std / np.sqrt(n))

# Compute degrees of freedom
df = n - 1  # Degrees of freedom

# Find critical t-value for left-tailed test
t_critical = stats.t.ppf(alpha, df)

# Compute p-value for left-tailed test
p_value = stats.t.cdf(t_statistic, df)

# Print results
print(f"T-Statistic: {t_statistic:.3f}")
print(f"Degrees of Freedom: {df}")
print(f"Critical T-Value (Left-Tailed): {t_critical:.3f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule
if t_statistic < t_critical:
    print("Reject the null hypothesis: The training program significantly reduced waiting times.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the training program reduced waiting times.")
```
