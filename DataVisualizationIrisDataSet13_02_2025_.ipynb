```python
#This assignment aims to perform data visualization on the Iris dataset using Matplotlib and Seaborn libraries. By exploring the dataset visually,
#students are expected to gain insights into the characteristics of the data and derive meaningful interpretations.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url="/content/Iris.csv"
data = pd.read_csv(url)
data.head()
```

```python
# @title Species

```

```python
# @title SepalLengthCm

from matplotlib import pyplot as plt
data['SepalLengthCm'].plot(kind='hist', bins=20, title='SepalLengthCm')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#1 What is the distribution of sepal lengths and widths among the Iris species?
data_sepallength = data.groupby(["Species", "SepalLengthCm"])["SepalLengthCm"].count()
data_sepalwidth  = data.groupby(["Species", "SepalWidthCm"])["SepalWidthCm"].count()
data_sepallength.plot(kind="bar", color="r", legend=True, position=0, width=0.25, label="Species Length")
#data_sepalwidth.plot(kind="bar", color="b", legend=True, position=0, width=0.25, label="Species width")


plt.rcParams["figure.figsize"] = (30, 20)

plt.xlabel("Species")
plt.ylabel("#Species Dimensions")
data_sepalwidth.plot(kind="bar", color="b", legend=True, position=0, width=0.25, label="Species width")
plt.rcParams["figure.figsize"] = (30, 20)

plt.xlabel("Species")
plt.ylabel("#Species Dimensions")
```

```python
#2. How does the petal length vary across different Iris species?
data_petalength = data.groupby(["Species", "PetalLengthCm"])["PetalLengthCm"].count()
data_petalength.plot(kind="bar", color="g", legend=True, position=0, width=0.25, label="Petal Length")
plt.rcParams["figure.figsize"] = (30, 20)

plt.xlabel("Species")
plt.ylabel("#Petal Length")
```

```python
#3 . Is there a relationship between sepal length and petal length?
sepal_length = data['SepalLengthCm']
petal_length = data['PetalLengthCm']

# Create a scatter plot
plt.scatter(range(len(sepal_length)),sepal_length, color='blue', label='Sepal Length')
plt.scatter(range(len(petal_length)),petal_length, color='red', label='Petal Length')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Length')
plt.title('Scatter Plot of Sepal and Petal Length')
plt.legend()

# Show the plot
plt.show()
```

```python

sns.relplot(x ="SepalLengthCm", y ="PetalLengthCm",
            hue ="Species", data = data);
```

```python
#4 How does the petal width vary across different Iris species?
# Load Iris dataset


# Create a box plot
sns.boxplot(data=data, x="Species", y="PetalWidthCm", palette="pastel")

plt.title("Box Plot of Petal Width Across Iris Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
#5 Are there any patterns observable in the pairwise relationships between features (sepal length, sepal width, petal length, petal width)?
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

# Load Iris dataset

df=pd.DataFrame(data)
df.head()
# Create pair plot
sns.pairplot(df, hue="Species", diag_kind="kde", palette="Set2")

plt.suptitle("Pairwise Relationships Between Features in Iris Dataset", y=1.02)
plt.show()
```

```python
#6  Can we distinguish between Iris species based on sepal and petal measurements?


# Scatter plot: Petal Length vs Petal Width
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=data,x="SepalLengthCm",y="SepalWidthCm",hue="Species",palette="Set1",s=100)
plt.title("Sepal Measurements")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.subplot(1, 2, 2)
sns.scatterplot(data=data,x="PetalLengthCm",y="PetalWidthCm",hue="Species",palette="Set2",s=100)
plt.title("Petal Measurements")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")

plt.tight_layout()
plt.show()
```
